## Softmax + Multiclass Regression

```python
import numpy as np  
import logging
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import re

class SoftmaxMulticlassClassification:
    def __init__(self):
        self.weights = None
        self.bias = None
    
    def softmax(self, z):
        z_shifted = z - np.max(z, axis=-1, keepdims=True)
        exp_z = np.exp(z_shifted)
        softmax_output = exp_z / np.sum(exp_z, axis=-1, keepdims=True)
        return softmax_output
    
    def reset_logging(self):
        # Clear existing handlers
        for handler in logging.root.handlers[:]:
            logging.root.removeHandler(handler)
    
    def draw_graph(self, log_file):
        iterations = []
        losses = []

        with open(log_file, "r") as f:
            for line in f:
                match = re.search(r"Iteration (\d+), Loss: ([\d\.]+)", line)
                if match:
                    iterations.append(int(match.group(1)))
                    losses.append(float(match.group(2)))

        
        plt.plot(iterations, losses, marker='o')
        plt.xlabel("Iteration")
        plt.ylabel("Loss")
        plt.title("Loss vs Iteration")
        plt.grid(True)
        plt.show()

    def train_bgd(self, X, y, log_file, learning_rate=0.01, n_iters=10000, graph = True):
        self.reset_logging()
        logging.basicConfig(filename=log_file, level=logging.INFO)
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))
        
        self.weights = np.random.rand(n_features, n_classes)
        self.bias = np.zeros(n_classes)

        y_one_hot = np.zeros((n_samples, n_classes))
        y_one_hot[np.arange(n_samples), y] = 1
        
        for i in range(n_iters):

            logits = np.dot(X, self.weights) + self.bias
            y_predicted = self.softmax(logits)
            loss = -np.mean(np.sum(y_one_hot * np.log(y_predicted), axis=1))
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y_one_hot))
            db = (1 / n_samples) * np.sum(y_predicted - y_one_hot, axis=0)
            self.weights -= learning_rate * dw
            self.bias -= learning_rate * db
            
            if i % 100 == 0:
                logging.info(f"Iteration {i}, Loss: {loss:.4f}")

        if graph:
            self.draw_graph(log_file)

        
    
    def train_sgd(self, X, y, log_file, learning_rate=0.01, n_iters=10000, graph = True):
        self.reset_logging()
        logging.basicConfig(filename=log_file, level=logging.INFO)
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))
        
        self.weights = np.random.rand(n_features, n_classes)
        self.bias = np.zeros(n_classes)
        
        for i in range(n_iters):
            for j in range(n_samples):
                logits = np.dot(X[j], self.weights) + self.bias
                y_predicted = self.softmax(logits)
                
                y_true = np.zeros(n_classes)
                y_true[y[j]] = 1
                
                loss = -np.log(y_predicted[y[j]] + 1e-9)
                
                dw = np.outer(X[j], (y_predicted - y_true))
                db = y_predicted - y_true
                
                self.weights -= learning_rate * dw
                self.bias -= learning_rate * db
            
            if i % 100 == 0:
                logging.info(f"Iteration {i}, Loss: {loss:.4f}")
        
        if graph:
            self.draw_graph(log_file)
    
    def predict(self, X):
        logits = np.dot(X, self.weights) + self.bias
        y_predicted = self.softmax(logits)
        y_predicted_class = np.argmax(y_predicted, axis=1)
        return y_predicted_class
    
    def evaluate_model(self, X_test, y_test):
        y_pred = self.predict(X_test)

        accuracy = np.mean(y_pred == y_test)
        print("Accuracy: {:.4f}".format(accuracy))
        
        classes = np.unique(np.concatenate((y_test, y_pred)))
        n_classes = len(classes)
        
        conf_matrix = np.zeros((n_classes, n_classes), dtype=int)
        for true_label, pred_label in zip(y_test, y_pred):
            conf_matrix[true_label, pred_label] += 1
        
        print("\nConfusion Matrix:")
        print(conf_matrix)
        
        print("\nClassification Report:")
        print("Class\tPrecision\tRecall\tF1-Score\tSupport")
        for cls in classes:
            tp = conf_matrix[cls, cls]
            fp = conf_matrix[:, cls].sum() - tp
            fn = conf_matrix[cls, :].sum() - tp
            precision = tp / (tp + fp) if (tp + fp) > 0 else 0
            recall = tp / (tp + fn) if (tp + fn) > 0 else 0
            f1_score = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0
            support = conf_matrix[cls, :].sum()
            print(f"{cls}\t{precision:.4f}\t\t{recall:.4f}\t{f1_score:.4f}\t\t{support}")
            
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# ------------------------------
# Assume X and y are already defined (e.g., loaded from your dataset)
# For example, X might be a numpy array with shape (N, 64, 64, 3)
# ------------------------------
img_width, img_height = 64, 64  
num_classes = 12

# # Reshape and convert X to channel-first format for PyTorch: (N, 3, 64, 64)
# X = X.reshape(-1, img_width, img_height, 3)
# X = np.transpose(X, (0, 3, 1, 2))
# X_test = X_test.reshape(-1, img_width, img_height, 3)
# X_test = np.transpose(X_test, (0, 3, 1, 2))
# # If y is one-hot encoded (shape: [N, num_classes]), convert it to class indices.
# if y_test.ndim == 2:
#     y_test = np.argmax(y_test, axis=1)

X_train_np, X_test_np, y_train_np, y_test_np = X, X_test, y, y_test

# ------------------------------
# Step 1: Data Augmentation and Preprocessing
# ------------------------------
# Create a custom dataset that converts the numpy image to a PIL image and applies transformations.
class CustomImageDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images  # images should be in (C, H, W) format
        self.labels = labels
        self.transform = transform
        
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image = self.images[idx]
        # If image is in channel-first format (3, H, W), transpose to (H, W, 3).
        if image.ndim == 3 and image.shape[0] == 3:
            image = image.transpose(1, 2, 0)
        image = transforms.ToPILImage()(image)
        if self.transform:
            image = self.transform(image)
        label = self.labels[idx]
        return image, label



# Define training transforms with augmentation and normalization.
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),  # Converts PIL Image to tensor in range [0, 1]
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Define validation/test transforms (only normalization).
val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Create dataset objects for training and validation.
train_dataset = CustomImageDataset(X_train_np, y_train_np, transform=train_transform)
val_dataset = CustomImageDataset(X_test_np, y_test_np, transform=val_transform)

batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# ------------------------------
# Step 2: Improved CNN Model and Regularization
# ------------------------------
# Define an improved CNN model with an additional convolutional block and updated dropout.
class ImprovedCNNModel(nn.Module):
    def __init__(self, num_classes):
        super(ImprovedCNNModel, self).__init__()
        self.features = nn.Sequential(
            # First Convolutional Block: 64x64 -> 32x32
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2),
            nn.Dropout(0.2),
            
            # Second Convolutional Block: 32x32 -> 16x16
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2),
            nn.Dropout(0.3),
            
            # Third Convolutional Block: 16x16 -> 8x8
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2),
            nn.Dropout(0.4),

            # Fourth Convolutional Block: 8x8 -> 4x4
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2),
            nn.Dropout(0.5)
        )
        
        # Global average pooling to reduce the spatial dimensions to 1x1.
        self.global_avg_pool = nn.AdaptiveAvgPool2d((1, 1))
        
        # Fully connected layers.
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)  # Final layer outputs raw logits.
        )
        
    def forward(self, x):
        x = self.features(x)
        x = self.global_avg_pool(x)
        x = self.classifier(x)
        return x

# Instantiate the model.
model = ImprovedCNNModel(num_classes)
print(model)

# Define optimizer (Adam with weight decay for L2 regularization) and loss function.
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
criterion = nn.CrossEntropyLoss()

# ------------------------------
# Training Loop
# ------------------------------
num_epochs = 20
history = {'train_loss': [], 'train_accuracy': [],
           'val_loss': [], 'val_accuracy': []}

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    total_train = 0

    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        running_corrects += torch.sum(preds == labels)
        total_train += inputs.size(0)
        
    epoch_loss = running_loss / total_train
    epoch_acc = running_corrects.double() / total_train
    history['train_loss'].append(epoch_loss)
    history['train_accuracy'].append(epoch_acc.item())
    
    # Validation phase.
    model.eval()
    val_loss = 0.0
    val_corrects = 0
    total_val = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_corrects += torch.sum(preds == labels)
            total_val += inputs.size(0)
            
    epoch_val_loss = val_loss / total_val
    epoch_val_acc = val_corrects.double() / total_val
    history['val_loss'].append(epoch_val_loss)
    history['val_accuracy'].append(epoch_val_acc.item())
    
    print(f'Epoch {epoch+1}/{num_epochs}, '
          f'Train Loss: {epoch_loss:.4f}, Train Acc: {epoch_acc:.4f}, '
          f'Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}')

# ------------------------------
# Updated Evaluation on Test/Validation Set with Thresholding
# ------------------------------
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in val_loader:
        outputs = model(inputs)
        # Convert logits to probabilities
        probs = torch.sigmoid(outputs)
        # Threshold probabilities: if probability > 0.3, predict presence (1), else absence (0)
        preds = (probs > 0.7).float()
        # Convert labels to one-hot if they are given as class indices
        if labels.ndim == 1:
            labels = torch.nn.functional.one_hot(labels, num_classes=num_classes).float()
        # Count element-wise correct predictions
        correct += torch.sum(preds == labels)
        total += labels.numel()
        

accuracy = correct / total
print(f"Test accuracy (using threshold 0.3): {accuracy.item() * 100:.2f}%")



# ------------------------------
# Plot the Training History
# ------------------------------
plt.figure(figsize=(12, 4))

# Accuracy plot.
plt.subplot(1, 2, 1)
plt.plot(history['train_accuracy'], label='Train Accuracy')
plt.plot(history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy over Epochs')

# Loss plot.
plt.subplot(1, 2, 2)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss over Epochs')

plt.tight_layout()
plt.show()
```
